```python
from datetime import datetime

def valida_data(data_str: str) -> bool:
    """
    Valida uma data no formato 'dd/mm/aaaa' e verifica se a pessoa tem 18 anos ou mais.
    Retorna True se for válida e idade >= 18, False caso contrário.
    """
    try:
        # Tenta converter a string em data
        dia, mes, ano = map(int, data_str.split('/'))
        data_nasc = datetime(ano, mes, dia).date()
        hoje = datetime.now().date()

        # Verifica se a data é futura
        if data_nasc > hoje:
            return False

        # Calcula a idade
        idade = hoje.year - data_nasc.year - ((hoje.month, hoje.day) < (data_nasc.month, data_nasc.day))

        # Retorna True apenas se idade >= 18
        return idade >= 18

    except ValueError:
        # Retorna False se a data for inválida (ex: 31/02/2020)
        return False


# Interação com o usuário
print(valida_data(input("Digite sua data de nascimento (dd/mm/aaaa): ")))
```
