```python
def menu_opcoes() -> int:
    """
    Exibe um menu de opções e valida a escolha do usuário.
    Retorna o número inteiro correspondente à opção escolhida (1, 2 ou 3).
    """
    while True:
        print("\nMENU DE OPÇÕES")
        print("1 - Cadastrar")
        print("2 - Exibir Frase")
        print("3 - Sair")
        
        try:
            opcao = int(input("Escolha uma opção (1 a 3): "))
            if 1 <= opcao <= 3:
                return opcao
            else:
                print("Opção inválida. Tente novamente.")
        except ValueError:
            print("Entrada inválida. Digite um número de 1 a 3.")
```

```python
escolha = menu_opcoes()
print(f"Você escolheu a opção {escolha}")
```
