```python
def valida_cpf(cpf: str) -> bool:
    """
    Valida um CPF no formato '999.999.999-99' conforme regras da Receita Federal.
    Retorna True se for válido, False caso contrário.
    """

    # Remove pontos e traço
    cpf = cpf.replace('.', '').replace('-', '')

    # Verifica se tem 11 dígitos e não é sequência repetida
    if len(cpf) != 11 or cpf == cpf[0] * 11:
        return False

    # Calcula o primeiro dígito verificador
    soma1 = sum(int(cpf[i]) * (10 - i) for i in range(9))
    resto1 = soma1 % 11
    dig1 = 0 if resto1 < 2 else 11 - resto1

    # Calcula o segundo dígito verificador (usa o primeiro dígito)
    soma2 = sum(int(cpf[i]) * (11 - i) for i in range(9)) + dig1 * 2
    resto2 = soma2 % 11
    dig2 = 0 if resto2 < 2 else 11 - resto2

    # Compara com os dígitos originais
    return cpf[-2:] == f"{dig1}{dig2}"


# Interação com o usuário
print(valida_cpf(input("Digite o CPF no formato 999.999.999-99: ")))
```
